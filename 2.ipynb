```python
import random

def create_random_matrix(n):
    limit = n * n * 10
    return [[random.randint(1, limit - 1) for _ in range(n)] for _ in range(n)]

def print_matrix(matrix, name):
    print(f"--- {name} ---")
    for row in matrix:
        for val in row:
            print(f"{val:5d}", end=" ")
        print()
    print("\n")

def solve_transpose():

    try:
        n = int(input("1보다 크고 5보다 작거나 같은 N을 입력하세요: "))
        if not (1 < n <= 5):
            print("2에서 5 사이의 숫자를 입력하세요.")
            return
    except ValueError:
        print("숫자만 입력 가능합니다.")
        return

    matrix_a = create_random_matrix(n)

    print_matrix(matrix_a, "Original Matrix A")

    transpose_a = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            transpose_a[j][i] = matrix_a[i][j]

    print_matrix(transpose_a, "Transposed Matrix A")

if __name__ == "__main__":
    solve_transpose()
```
