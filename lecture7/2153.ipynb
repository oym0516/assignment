```python
def is_prime(n):
    if n == 1:
        return True
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
        return True

x = list(input())
cnt = 0
for i in x:
    if i.isupper():
        cnt += (ord(i) - 64)
    else:
        cnt += (ord(i) - 96)

if is_prime(cnt):
    print('It is a prime word.')
else:
    print('It is not a prime word.')
```
