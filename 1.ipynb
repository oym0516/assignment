```python
import random

def create_random_matrix(n):
    limit = n * n * 10
    return [[random.randint(1, limit - 1) for _ in range(n)] for _ in range(n)]

def print_matrix(matrix, name):
    print(f"--- {name} ---")
    for row in matrix:
        for val in row:
            print(f"{val:5d}", end=" ")
        print()
    print("\n")

def solve_assignment():
    try:
        n = int(input("1보다 크고 5보다 작거나 같은 N을 입력하세요: "))
        if not (1 < n <= 5):
            print("입력 범위 오류! (2, 3, 4, 5 중 하나를 입력하세요)")
            return
    except ValueError:
        print("숫자만 입력해 주어야 합니다.")
        return

    A = create_random_matrix(n)
    B = create_random_matrix(n)
    C = create_random_matrix(n)

    print_matrix(A, "Matrix A")
    print_matrix(B, "Matrix B")
    print_matrix(C, "Matrix C")


    AxB = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                AxB[i][j] += A[i][k] * B[k][j]

    result = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            result[i][j] = AxB[i][j] + C[i][j]

    print_matrix(result, "Result (A x B + C)")

if __name__ == "__main__":
    solve_assignment()
```
